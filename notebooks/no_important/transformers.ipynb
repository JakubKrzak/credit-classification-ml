```python
class Drop(BaseEstimator,TransformerMixin):
    def __init__(self, columns_to_drop=None):
        self.columns_to_drop = columns_to_drop
        
    def fit(self, X,y=None):
        return self
    
    def transform(self, X):
        
        X = X.drop(columns= self.columns_to_drop, errors='ignore')
        
        return X
    
    # Column in this transformer
    # 1.ID
    # 2.Customer_ID
    # 3.Name
    # 4.SSN
```

```python
class ToNumRemoveChars(BaseEstimator,TransformerMixin):
    def __init__(self, chars_to_remove=r'[^0-9.,]+'):
        self.chars_to_remove = chars_to_remove
        
    
    def fit(self, X,y=None):
        return self
    
    def transform(self, X):
    
       
        df = pd.DataFrame(X).copy()
       
        for col in df.columns:
            s = df[col].astype(str)
            s = s.str.replace(self.chars_to_remove, '', regex=True)
            s = s.str.replace(',', '.', regex=False)
            df[col] = pd.to_numeric(s, errors='coerce')
            
        
        return df
    
    # Column in this transformer
    # 1.Age
    # 2.Annual_Income
    # 3.Monthly_Inhand_Salary 
    # 4.Num_Bank_Accounts 
    # 5.Num_Credit_Card
    # 6.Interest_Rate
    # 7.Num_of_Loan
    # 8.Delay_from_due_date 
    # 9.Num_of_Delayed_Payment 
    # 10.Changed_Credit_Limit
    # 11.Num_Credit_Inquiries
    # 12.Outstanding_Debt 
    # 13.Credit_Utilization_Ratio 
    # 14.Total_EMI_per_month
    # 15.Amount_invested_monthly 
    # 16.Monthly_Balance
    
```

Numeryczne

```python
class AgeTransformer(BaseEstimator,TransformerMixin):
    def __init__(self):
        self.min_age_ = 18
        self.max_age_ = 100
        pass
    
    def fit(self,X,y=None):
        return self
    
    def transform(self,X):
        
        X_df = pd.DataFrame(X).copy()
        
        col = X_df.columns[0]
        
        X_df.loc[(X_df[col] < self.min_age_) | (X_df[col] > self.max_age_), col] = np.nan
        
        return X_df
    
# Column in this transformer
# 1.Age
```

```python
class ValueCliperq99(BaseEstimator,TransformerMixin):
    def __init__(self, q_max=0.99):
        self.q_max = q_max
        self.borders_ = {}
        
    
    def fit(self,X,y=None):
        X_df = pd.DataFrame(X).copy()
        
         #s≈Çownik na progi
        
        
        for col in X_df.columns:
            data = pd.to_numeric(X_df[col], errors='coerce') #przejscie po kazdej kolumnie podanej do transformatora
            self.borders_[col] = data.quantile(self.q_max) #zapisuje prog dla kazdej kategori w slowniku
            
            
            
        return self
    
    def transform(self,X):
        X_df = pd.DataFrame(X).copy()
        
        for col in X_df.columns:
            X_df.loc[X_df[col] <= 0, col] = np.nan
        
            col_border = self.borders_[col]
        
            X_df.loc[(X_df[col] > col_border), col] = np.nan
        
        return X_df
    
    # Column in this transformer
    # 1.AnnualIncome
    # 2.MonthlyInhandSalaryTransform
    # 3.delay_from_due_date
    # 4.num_of_delayed_payment
    # 5.num_credit_inquiries
    # 6.credit_utilization_ratio
    
    
```

```python
class ValueClipperq90(BaseEstimator,TransformerMixin):
    def __init__(self, q_max=0.90):
        self.q_max = q_max
        self.borders_ = None
        
    
    def fit(self,X,y=None):
        X_df = pd.DataFrame(X).copy()
        
        self.borders_ = {}
        
        for col in X_df.columns:
            data = pd.to_numeric(X_df[col], errors='coerce')
            self.borders_[col] = data.quantile(self.q_max)   
        return self
    
    def transform(self,X):
        X_df = pd.DataFrame(X).copy()
        
        for col in X_df.columns:
            X_df.loc[X_df[col] <= 0, col] = np.nan
            
            col_border = self.borders_[col]
            
            X_df.loc[(X_df[col] > col_border), col] = np.nan
            
        return X_df
    # Column in this transforme
    # 1.num_bank_accounts
    # 2.clean_num_credit_card
    # 3.clean_intrest_rate
    # 4.amount_invested_monthly
```

```python
class EmiPerMonthTransformer(BaseEstimator,TransformerMixin):
    def __init__(self, q_max=0.94):
        self.q_max = q_max
        self.quant_ = None
        
    def fit(self,X,y=None):
        X_df = pd.DataFrame(X).copy()
        
        col = X_df.columns[0]
        
        data = pd.to_numeric(X_df[col], errors='coerce')
        
        
        self.quant_ = data.quantile(self.q_max)
        
        return self
    
    def transform(self, X):
        X_df = pd.DataFrame(X).copy()
        
        col = X_df.columns[0]
        
        data = pd.to_numeric(X_df[col], errors='coerce')
        
        X_df.loc[(data < 0), col] = np.nan
        
        X_df.loc[(data > self.quant_), col] = np.nan
        
        return X_df
    
#1.total_emi_per_month
```

```python
class MonthlyBalance(BaseEstimator,TransformerMixin):
    def __init__(self):
        pass
    
    def fit(self,X,y=None):
        return self
    
    def transform(self,X):
        X_df = pd.DataFrame(X).copy()
        
        col = X_df.columns[0]
        
        data = pd.to_numeric(X_df[col], errors='coerce')
        
        X_df.loc[(data <= 0), col] = np.nan
        
        return X_df
    #Column in this transformer
    #1.MonthlyBalance
```

Kategoryczne

```python
class TypeOfLoan(BaseEstimator,TransformerMixin):
    def __init__(self, chars_to_remove=r'and|\s|,'):
       self.chars_to_remove = chars_to_remove
    
    def fit(self,X,y=None):
        return self
    
    def transform(self,X):
        X_df = pd.DataFrame(X).copy()
        
        col = X_df.columns[0]
        
        loan_series = X_df[col].fillna("")
            
        loan_series = ( #czysczenie separatorow i zastapienie ich przecinkiem
            loan_series.astype(str)
            .str.replace(self.chars_to_remove, ',', regex=True)
            .str.strip()
        )
        
        loan_series = loan_series.apply(
            lambda x: [] if x == "" else [t.strip() for t in x.split(',') if t.strip()]
        )
        
        X_df[col] = loan_series
        return X_df
    
    # Column in this transformer
    #1.Type_of_Loan
```

```python
class CreditMix(BaseEstimator,TransformerMixin):
    def __init__(self, placeholder='_'):
        self.placeholder = placeholder
        
    def fit(self,X,y=None):
            return self
        
    def transform(self, X):
            
        X_df = pd.DataFrame(X).copy()
            
        col = X_df.columns[0]
            
        credit_series = X_df[col]
            
        credit_series = (
            credit_series.astype(str)
            .str.replace(self.placeholder, '', regex=False)
        )
            
            
        X_df[col] = credit_series
        return X_df
    
    # Column in this transformer
    #1.Credit_Mix'
```

```python
class CreditHistoryAge(BaseEstimator,TransformerMixin):
    def __init__(self, extract=r'(\d+)\s*Years?(?:.*?(\d+)\s*Months?)?'):
        self.extract = extract

    def fit(self,X,y=None):
        return self

    def transform(self, X):
        X_df = pd.DataFrame(X).copy()
        col = X_df.columns[0]

        history = (
            X_df[col]
            .astype(str)
            .str.extract(self.extract, expand=True)
            .fillna(0)
            .astype(float)
        )

        months = history.iloc[:, 0] * 12 + history.iloc[:, 1]
        months = months.mask(months <= 0)

        X_df[col] = months
        return X_df
    # Column in this transformer
    #1.Credit_History_Age
    # Column in this transformer
    #1.Credit_History_Age
```

```python
class PaymentOfMinAmount(BaseEstimator,TransformerMixin):
    def __init__(self, placeholder='NM'):
        self.placeholder = placeholder
    
    def fit(self,X,y=None):
        return self
        
    def transform(self, X):
        X_df = pd.DataFrame(X).copy()
        
        col = X_df.columns[0]
        
        payment = (
            X_df[col]
            .astype(str)
            .str.replace(self.placeholder, '', regex=False)
            )
        
        X_df[col] = payment
        return X_df
    
    # Column in this transformer
    #1.Payment_of_Min_Amount
```

```python
class Behavior(BaseEstimator,TransformerMixin):
    def __init__(self, remove='!@9#%8'):
        self.remove = remove
    
    def fit(self,X,y=None):
        return self
        
    def transform(self, X):
        X_df = pd.DataFrame(X).copy()
        
        col = X_df.columns[0]
        
        behaviour = (
            X_df[col]
            .astype(str)
            .str.replace(self.remove, '', regex=False)
        )
        X_df[col] = behaviour
        return X_df
    # Column in this transformer
    #1.Behaviour
```

```python
class Ocupation(BaseEstimator,TransformerMixin):
    def __init__(self, remove='_______'):
        self.remove = remove
    
    def fit(self,X,y=None):
        return self
        
    def transform(self, X):
        X_df = pd.DataFrame(X).copy()
        
        col = X_df.columns[0]
        
        occupation = (
            X_df[col]
            .astype(str)
            .str.replace(self.remove, '', regex=False)
        )
        
        X_df[col] = occupation
        return X_df
    # Column in this transformer
    #1.Occupation
```

```python
def fit_transform(df):
    df = df.copy()

    # 1. Drop kolumn
    df = df.drop(columns=c_drop, errors="ignore")

    # 2. ToNumRemoveChars na kolumnach numerycznych z dziwnymi znakami
    to_num = ToNumRemoveChars()
    df[c_ToNumRemoveChars] = to_num.fit_transform(df[c_ToNumRemoveChars])

     #3. Age ‚Äì je≈õli masz osobny transformer (np. Age), inaczej mo≈ºesz to na razie pominƒÖƒá
    age_tr = AgeTransformer()
    df[c_age] = age_tr.fit_transform(df[c_age])

    # 4. ValueCliperq99
    vc99 = ValueCliperq99()
    df[c_ValueCliperq99] = vc99.fit_transform(df[c_ValueCliperq99])

    # 5. ValueClipperq90
    vc90 = ValueClipperq90()
    df[c_ValueClipperq90] = vc90.fit_transform(df[c_ValueClipperq90])

    # 6. Total_EMI_per_month
    emi_tr = EmiPerMonthTransformer()
    df[c_EmiPerMonthTransformer] = emi_tr.fit_transform(df[c_EmiPerMonthTransformer])

    # 7. Monthly_Balance
    mb_tr = MonthlyBalance()
    df[c_MonthlyBalance] = mb_tr.fit_transform(df[c_MonthlyBalance])

    # 8. Type_of_Loan
    tol_tr = TypeOfLoan()
    df[c_TypeOfLoan] = tol_tr.fit_transform(df[c_TypeOfLoan])

    # 9. Credit_Mix
    cm_tr = CreditMix()
    df[c_CreditMix] = cm_tr.fit_transform(df[c_CreditMix])

    # 10. Credit_History_Age
    cha_tr = CreditHistoryAge()
    df[c_CreditHistoryAge] = cha_tr.fit_transform(df[c_CreditHistoryAge])


    # 11. Payment_of_Min_Amount
    pmin_tr = PaymentOfMinAmount()
    df[c_PaymentOfMinAmount] = pmin_tr.fit_transform(df[c_PaymentOfMinAmount])

    # 12. Payment_Behaviour
    beh_tr = Behavior()
    df[c_Behavior] = beh_tr.fit_transform(df[c_Behavior])

    # 13. Occupation
    occ_tr = Ocupation()
    df[c_Ocupation] = occ_tr.fit_transform(df[c_Ocupation])

    return df
```

```python
df = fit_transform(df)
```
